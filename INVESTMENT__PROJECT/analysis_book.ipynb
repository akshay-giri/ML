```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
X_LE = LabelEncoder()
Y_LE = LabelEncoder()
SS = StandardScaler()

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("investing_program_prediction_data.csv")
df.head()
```

```python
for column in ['PE' + str(i) for i in range(1,16)]:
    df[column] = df[column].apply(lambda x:x[1]).astype(np.int)
```

```python
df
```

```python
df.info()
```

```python
X = df.drop("InvType",axis=1)
X['SE2'] = X_LE.fit_transform(X['SE2'])
X = SS.fit_transform(X)
X
```

```python
df.InvType.value_counts()
```

```python
Y = Y_LE.fit_transform(df.InvType)
Y
```

```python
Y_LE.inverse_transform([0])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_pred))
```

```python
pickle.dump(X_LE, open("X_LE.pkl",'wb'))
pickle.dump(Y_LE, open("Y_LE.pkl",'wb'))
pickle.dump(SS, open("SS.pkl",'wb'))

pickle.dump(model, open("GBC_model.pkl",'wb'))
```
