```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1

df=df[~((df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR))).any(axis=1)]
df
```

```python
from sklearn.model_selection import train_test_split

X=df[['Avg. Session Length','Time on App','Time on Website', 'Length of Membership']]
y=df['Yearly Amount Spent']

X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()

lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
model_report = pd.DataFrame()
tmp1 = pd.Series({'Model': " Linear Regression ",
                 'R2 Score' : r2_score(y_test,y_pred),
                 'MAE': mean_absolute_error(y_test,y_pred),
                 'MSE': mean_squared_error(y_test,y_pred)})

model_lr_report = model_report.append(tmp1, ignore_index = True)
model_lr_report
```

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
l1=Lasso()
l1.fit(X_train,y_train)
ypred=l1.predict(X_test)
```

```python
model_report = pd.DataFrame()
tmp1 = pd.Series({'Model': " Lasso ",
                 'R2 Score' : r2_score(y_test,ypred),
                 'MAE': mean_absolute_error(y_test,ypred),
                 'MSE': mean_squared_error(y_test,ypred)})

model_l1_report = model_report.append(tmp1, ignore_index = True)
model_l1_report
```

```python
l2=Ridge()
l2.fit(X_train,y_train)
ypred2=l2.predict(X_test)
```

```python
model_report = pd.DataFrame()
tmp1 = pd.Series({'Model': " Ridge ",
                 'R2 Score' : r2_score(y_test,ypred2),
                 'MAE': mean_absolute_error(y_test,ypred2),
                 'MSE': mean_squared_error(y_test,ypred2)})

model_l2_report = model_report.append(tmp1, ignore_index = True)
model_l2_report
```

```python
cols = ['Model', 'R2 Score', 'MAE', 'MSE']
model = pd.DataFrame(columns = cols)
model = model.append([model_lr_report,model_l1_report,model_l2_report], ignore_index = True)
model
```
